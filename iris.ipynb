```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('dataset.xlsx')
```

```python
df
```

```python
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder()
```

```python
values=encoder.fit_transform(df[['iris type']])
```

```python
values
```

```python
df['iris type'] = values
```

```python
df
```

```python
train_df = df[['sepal length','sepal width','petal length','petal width']]
train_df
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(train_df,df['iris type'],test_size = 0.33)
```

```python
y_train
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
model.predict(np.array([1,2,3,1]).reshape(1,-1))
```

```python
x_test
```

```python
from sklearn.metrics import f1_score
y_predict = model.predict(x_test)
score = f1_score(y_test,y_predict,average = 'micro')
```

```python
score
```

```python
#now let's do it with a RFC tuned with gridsearch
from sklearn.model_selection import GridSearchCV
RFC = RandomForestClassifier(random_state=42)
param_grid = { 
    'n_estimators': [20, 100],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth' : [2,3,4,5,6,7],
    'criterion' :['gini', 'entropy']
}
tunedRFC = GridSearchCV(RFC,param_grid=param_grid,cv=5)
tunedRFC.fit(x_train,y_train)
```

```python
tunedRFC.best_params_
```

```python
finalRFC = RandomForestClassifier(criterion='gini',max_depth=4,max_features='auto',n_estimators=100)
finalRFC.fit(x_train,y_train)
y_predict_tuned = finalRFC.predict(x_test)
score_tuned = f1_score(y_test,y_predict_tuned,average='micro')
score_tuned
```
